```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#for loading the file we need to put .csv to make it readable
haberman  = pd.read_csv("haberman.csv")
```

```python
#no of coloumns and features
print(haberman.columns)
#in this file only dat are aviable Name are give by me
```

```python
haberman["sur_stat"].value_counts
```

```python
haberman.plot(kind='scatter',x='Age',y='Op_year');
plt.show()
```

```python
#2d scatter plot
sns.set_style("whitegrid");
sns.FacetGrid(haberman,hue="sur_stat",size=4) \
   .map(plt.scatter,"Age","Op_year") \
   .add_legend();
plt.show();
```

```python
#3d scatter plot
plt.close();
sns.set_style("whitegrid");
sns.pairplot(haberman,hue="sur_stat",size=3);
plt.show();
```

```python
#boxplot on haberman
sns.boxplot(x="Age",y="Op_year",data=haberman)
plt.show()
```

```python
#violin plot on haberman
sns.violinplot(x="Age",y="Op_year",data=haberman,size=990999999999900000000000000000000000)
plt.show()
#i dont know wat value to give on size so i gave randdom values
```

```python
sns.FacetGrid(haberman,hue="sur_stat",size=5) \
   .map(sns.distplot,"Age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(haberman,hue="sur_stat",size=5) \
   .map(sns.distplot,"Op_year") \
   .add_legend();
plt.show();
```

```python
counts, bin_edges =np.histogram(haberman['sur_stat'],bins=10,
                               density=True)


pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts, bin_edges= np.histogram(haberman['sur_stat'],bins=20,
                                density=True)
pdf=counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);


plt.show();
```
